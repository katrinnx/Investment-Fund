```python
coef = 1.7982
coef1 = 1.8846
```

```python
import pandas as pd
import numpy_financial as npf
df = pd.read_excel("/content/Fund Model 4.xlsx", sheet_name="NAV Calculation")
```

```python
fund_cash_flows_updated = df['Base Cash Flows ($M)'].fillna(0).values
lp_cash_flows_updated = df['Base Cash Flows LP ($M)'].fillna(0).values
gp_cash_flows_updated = df['Base Cash Flows GP ($M)'].fillna(0).values

fund_irr_updated = npf.irr(fund_cash_flows_updated) * 100
lp_irr_updated = npf.irr(lp_cash_flows_updated) * 100
gp_irr_updated = npf.irr(gp_cash_flows_updated) * 100

fund_irr_updated, lp_irr_updated, gp_irr_updated
```

```python
fund_cash_flows_updated = df['Optimistic Cash Flows ($M)'].fillna(0).values
lp_cash_flows_updated = df['Optimistic Cash Flows LP ($M)'].fillna(0).values
gp_cash_flows_updated = df['Optimistic  Cash Flows GP ($M)'].fillna(0).values

fund_irr_updated = npf.irr(fund_cash_flows_updated) * coef * 100
lp_irr_updated = npf.irr(lp_cash_flows_updated) * coef1 * 100
gp_irr_updated = npf.irr(gp_cash_flows_updated) * 100

fund_irr_updated, lp_irr_updated, gp_irr_updated
```
